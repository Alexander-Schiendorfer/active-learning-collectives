```python
import numpy as np

plants = [10]
sps = [5,10, 25]

for plant in plants:
    for sp in sps:
        fileName = "plants_"+str(plant)+"_avpp_0_sps_"+ str(sp) + "_costs.csv"
        data = np.genfromtxt(fileName, delimiter=";")
        plot(data[:,0], data[:,1], label=str(sp) + " sps")
        title(str(plant) + " plants")
        legend(loc=2)
    figure()
```

```python

#	protected int[] powerPlants = {10};
#	protected int[] samplingPoints = {5, 10, 25};
 
plants = [10]
sps = [5, 10, 25]
objectives = ["costs", "pdelta-pos", "pdelta-neg"]

for obj in objectives:
    for plant in plants:
        for sp in sps:
            fileName = "plants_"+str(plant)+"_avpp_0_sps_"+ str(sp) + "_"+obj+".csv"
            data = np.genfromtxt(fileName, delimiter=";")
            plot(data[:,0], data[:,1], label=str(sp) + " sps")
            title(obj + " " +str(plant) + " plants")
            legend(loc=2)
        figure()
    figure()
```
